```python
import numpy as np

dropout_ratio = 0.6
x = np.ones(10)
print(x)
mask = np.random.rand(10) > dropout_ratio
print(mask)
y = x * mask
print(y)
```

```python
# 테스트 시
scale = 1 - dropout_ratio
y = x * scale
print(y)
```

## 역 드롭아웃

```python
# 학습 시
scale = 1 - dropout_ratio
print(scale)
mask = np.random.rand(*x.shape) > dropout_ratio
y = x * mask / scale
print(y)
# 테스트
y = x
print(y)
```
